```python
df_ = pd.read_csv('/zine/HPC02S1/ex-dveloza/AGORA/apps/fred_colombia_implementation/process_inputs/output_files/camas_uci_colombia.csv')
```

```python
departamentos = ["08 - Atlántico",
               "11 - Bogotá, D.C.",
               "13 - Bolívar",
               "15 - Boyacá",
               "20 - Cesar",
               "05 - Antioquia",
               "25 - Cundinamarca",
               "19 - Cauca",
               "23 - Córdoba",
               "27 - Chocó",
               "17 - Caldas",
               "18 - Caquetá",
               "86 - Putumayo",
               "47 - Magdalena",
               "52 - Nariño",
               "50 - Meta",
               "73 - Tolima",
               "85 - Casanare",
               "68 - Santander",
               "41 - Huila",
               "94 - Guainía",
               "63 - Quindio",
               "70 - Sucre",
               "76 - Valle del Cauca",
               "66 - Risaralda",
               "44 - La Guajira",
               "54 - Norte de Santander",
               "88 - Archipiélago de San Andrés, Providencia y Santa Catalina",
               "81 - Arauca",
               "91 - Amazonas",
               "99 - Vichada",
               "95 - Guaviare",
               "97 - Vaupés"]
```

```python
df_data = df_[(df_.Departamento.isin(departamentos)) & (~df_.Municipio.astype(str).str.contains('total'))]
```

```python
dicc_names = {'Fecha' : 'date',
              'Capacidad' : 'capacity', 
              'Departamento' : 'department', 
              'COVID_confirmados' : 'covid_confirmed', 
              'COVID_sospechosos' : 'covid_suspected', 
              'No_COVID' : 'non_covid', 
              'Total_camas' : 'total_beds',
              'Camas_disponibles' : 'available_beds'}

df_processed = df_data[['Fecha', 'Capacidad', 'Departamento', 'COVID_confirmados', 'COVID_sospechosos', 'No_COVID', 'Total_camas',
       'Camas_disponibles']].groupby(by=['Fecha', 'Capacidad', 'Departamento']).sum().reset_index().rename(columns=dicc_names)

df_processed['date'] = pd.to_datetime(df_processed['date'])
```

```python
df_processed.to_csv('/zine/HPC02S1/ex-dveloza/AGORA/apps/fred_colombia_implementation/process_inputs/output_files/UCI/camas_uci_departamentos_colombia.csv', index=False)
```
